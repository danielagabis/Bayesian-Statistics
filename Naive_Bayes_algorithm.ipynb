**Naive Bayes algorithm**                                             

This simulation shows the Naive Bayes algorithm for classifying whether the car is stolen based on its features.


The dataset includes two features:

• “Yes” (the red SUV domestic is stolen)

• “No” (the red SUV domestic is not stolen)


The model was taught using Categorical Naive Bayes, which calculates probabilities based on the frequency of specific labels within a dataset.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.naive_bayes import CategoricalNB

# Dataset
data = {
    'Color': ['Red', 'Red', 'Red', 'Yellow', 'Yellow', 'Yellow', 'Yellow', 'Yellow', 'Red', 'Red'],
    'Type': ['Sports', 'Sports', 'Sports', 'Sports', 'Sports', 'SUV', 'SUV', 'SUV', 'SUV', 'Sports'],
    'Origin': ['Domestic', 'Domestic', 'Domestic', 'Domestic', 'Imported', 'Imported', 'Imported', 'Domestic', 'Imported', 'Imported'],
    'Stolen': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes']
}

df = pd.DataFrame(data)

# Preprocessing
feature_encoder = OrdinalEncoder()
X = feature_encoder.fit_transform(df[['Color', 'Type', 'Origin']])
y = df['Stolen']

# Training the Model
model = CategoricalNB(alpha=1e-10)
model.fit(X, y)

# Simulation
test_case = pd.DataFrame([['Red', 'SUV', 'Domestic']], columns=['Color', 'Type', 'Origin'])
test_encoded = feature_encoder.transform(test_case)

# Predict Probability and Class
probs = model.predict_proba(test_encoded)
prediction = model.predict(test_encoded)

# Output Results
print("--- Simulation Results ---")
print(f"Target: Red SUV Domestic")
print(f"Probability [No]:  {probs[0][0]:.3f}")
print(f"Probability [Yes]: {probs[0][1]:.3f}")
print(f"Final Decision: Car is Stolen? -> {prediction[0]}")
```
